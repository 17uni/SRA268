### Minimum Temp Compared to Snowfall

#### Outline

The purpose of this comparison is to compare the minimum temperatures to the amount of snowfall. The comparison looks at all the days of 2017 where temperatures were freezing (32 degrees fahrenheit) or lower in State College, PA. Days with zero snow fall were included in the data set, as were days where the temperature min was 32 degrees fahrenheit.
